Again let's look at the percentage of Nan values in the different columns. It may be usefull later.

```python
drop_advocate = []
for i,attribute in enumerate(beers_advocate.columns):
    nan_ratings = beers_advocate[beers_advocate[attribute].isna()]
    percentage = len(nan_ratings)/len(beers_advocate)*100
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_advocate),"->",percentage,"%")
    if percentage>20:  #Value to change for column deletion (used later)
        drop_advocate.append(attribute)
```

```python
drop_ratebeer =[]
for i,attribute in enumerate(beers_ratebeer.columns):
    nan_ratings = beers_ratebeer[beers_ratebeer[attribute].isna()]
    percentage = len(nan_ratings)/len(beers_ratebeer)*100
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_ratebeer),"->",percentage,"%")
    if percentage>20:#Value to change for column deletion (used later)
        drop_ratebeer.append(attribute)
```

```python
drop_matched = []
for i,attribute in enumerate(beers_matched.columns):
    nan_ratings = beers_matched[beers_matched[attribute].isna()]
    percentage = len(nan_ratings)/len(beers_matched)*100
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_matched),"->",percentage,"%")
    if percentage>20: #Value to change for column deletion (used later)
        drop_matched.append(attribute)
```

Since ba_score, bros_score, z_score, avg_valid_ratings_matched,overall_score, style_score  have a high percentage of Nan Values we decided to delete the full column. If needed we may recompute them later for our own needs.

```python
#Comment out for your needs. One sets the columns to delete, the other takes the columns to delete based on a threshold, it was computed in the cell above. 
# The lines with the threshold have a comment. The value is quite low but in the end we never used those cleaned files. Instead we chose in merge_into_onedataset which columns to use or not.



#drop_advocate = ['ba_score','bros_score','zscore','avg_matched_valid_ratings']
#drop_ratebeer = ['overall_score','zscore','avg_matched_valid_ratings', 'style_score']
#drop_matched = ['ba_score','bros_score','zscore','avg_matched_valid_ratings', 'style_score', 'overall_score','zscore.1','avg_matched_valid_ratings.1']
#We can use drop matched with added columns because we droped avg_matched_valid_ratings and zscore in both datasets so droping it here would be an option. (But t his can be later changed by uncommenting the previous line).
beers_advocate_clean = beers_advocate.drop(columns=drop_advocate)
beers_matched_clean = beers_matched.drop(columns=drop_matched)
beers_ratebeer_clean = beers_ratebeer.drop(columns=drop_ratebeer)
```

```python
for i,attribute in enumerate(beers_advocate_clean.columns):
    nan_ratings = beers_advocate_clean[beers_advocate_clean[attribute].isna()]
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_advocate_clean),"->",len(nan_ratings)/len(beers_advocate_clean)*100,"%")
for i,attribute in enumerate(beers_ratebeer_clean.columns):
    nan_ratings = beers_ratebeer_clean[beers_ratebeer_clean[attribute].isna()]
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_ratebeer_clean),"->",len(nan_ratings)/len(beers_ratebeer_clean)*100,"%")
for i,attribute in enumerate(beers_matched_clean.columns):
    nan_ratings = beers_matched_clean[beers_matched_clean[attribute].isna()]
    print("Number of Nan values for ", attribute, ":", len(nan_ratings),"/",len(beers_matched_clean),"->",len(nan_ratings)/len(beers_matched_clean)*100,"%")
```

For the needs of our project we need beers with a certain amount of ratings. For now we just want to keep it at 0. We will also delete the ones with a smaller number of rating than the threshold. 

```python
beers_advocate_clean = beers_advocate_clean[beers_advocate_clean['nbr_ratings'] > 0.5]
beers_matched_clean = beers_matched_clean[(beers_matched_clean['nbr_ratings'] > 0.5) | (beers_matched_clean['nbr_ratings.1']>0.5)]
beers_ratebeer_clean = beers_ratebeer_clean[beers_ratebeer_clean['nbr_ratings'] > 0.5]
```

```python
#Save the nee files
beers_advocate_clean.to_csv(os.path.join(ADVOCATE_PATH,'beers_cleaned.csv'),index=False)
beers_matched_clean.to_csv(os.path.join(MATCHED_PATH,'beers_cleaned.csv'),index=False)
beers_ratebeer_clean.to_csv(os.path.join(RATEBEER_PATH,'beers_cleaned.csv'),index=False)
```
